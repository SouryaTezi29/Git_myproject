```python
def check_divisibility_all(lst, divisors):
    return all(all(n % divisor == 0 for divisor in divisors) for n in lst)

lst = [14, 12, 4, 18]
divisors = [2, 7]
print(check_divisibility_all(lst, divisors))  # True
```

```python
def check_divisibility_all(list1, list2):
    divisible_numbers = []

    for num2 in list2:
        if all(num2 % num1 == 0 for num1 in list1):
            divisible_numbers.append(num2)

    return divisible_numbers

# Create two lists
list1 = [2, 3, 5]
list2 = [10, 12, 15, 20, 30]

# Check for divisibility
divisible_numbers = check_divisibility_all(list1, list2)

# Print the numbers
print("Numbers in list2 divisible by all numbers in list1:")
for num in divisible_numbers:
    print(num)
```
